```python
cols_to_drop=["uniq_id","crawl_timestamp","image","is_FK_Advantage_product","product_rating","overall_rating","brand","product_specifications"]
```

```python
product=df.drop(cols_to_drop,axis=1)

product.head()
```

```python
len(product)
```

```python
product = product.drop_duplicates()

len(product)
```

```python
product.isna().sum()
```

```python
product=product.dropna()
```

```python
len(product)
```

```python
product.head()
```

```python
product.isna().sum()
```

```python
product.to_csv('product.csv', index=False)
```

```python
model_train_dataset=product.drop(["product_url","product_name","retail_price","discounted_price"],axis=1)
```

```python
model_train_dataset.head()
```

```python
model_train_dataset['Merged']=model_train_dataset.groupby('product_category_tree')['pid'].transform(lambda x: ', '.join(x))
```

```python
model_train_dataset1 = model_train_dataset.drop_duplicates(subset='product_category_tree')[['product_category_tree', 'Merged']].reset_index(drop=True)
```

```python
model_train_dataset1
```

```python
rename_cols={"product_category_tree":"description","Merged":"product_id"}
```

```python
model_train_dataset1.rename(columns=rename_cols,inplace=True)
```

```python
model_train_dataset1.to_csv('model_dataset.csv', index=False)
```
